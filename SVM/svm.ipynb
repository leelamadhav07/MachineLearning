```python
import pandas as pd
from inspect import Parameter
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_features=2, n_classes=2, n_redundant=0,)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear')
svc.fit(x_train, y_train)
svc.coef_
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

print("classification report :",classification_report(y_test,y_pred))
print("confusion matrix : ",confusion_matrix(y_test,y_pred))
print("Accuracy score :",accuracy_score(y_test,y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(x_train, y_train)
y_pred1 = rbf.predict(x_test)
```

```python
print("classification report :",classification_report(y_test,y_pred1))
print("confusion matrix : ",confusion_matrix(y_test,y_pred1))
print("Accuracy score :",accuracy_score(y_test,y_pred1))
```

```python
poly = SVC(kernel='poly')
poly.fit(x_train,y_train)
y_pred2 = poly.predict(x_test)

print("classification report :",classification_report(y_test,y_pred2))
print("confusion matrix : ",confusion_matrix(y_test,y_pred2))
print("Accuracy score :",accuracy_score(y_test,y_pred2))
```

```python
# HyperParameter Tuning
from sklearn.model_selection import GridSearchCV

# defining parameters range 
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}

grid = GridSearchCV(SVC(), param_grid = param_grid, refit=True, cv = 5,verbose=3)
grid.fit(x_train, y_train)

grid.best_params_
```

```python
y_pred4 = grid.predict(x_test)
print("classification report :",classification_report(y_test,y_pred4))
print("confusion matrix :",confusion_matrix(y_test,y_pred4))
```
